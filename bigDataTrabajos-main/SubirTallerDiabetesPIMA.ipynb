# Importar las librerias Necesarias

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
```

# Cargar datos 

```python
datos=np.loadtxt('datos.csv',delimiter=',')
```

# Mostrar Datos

```python

print(datos[:,:8])
```

# Total de Registros

```python
print(len(datos))
```

# Mostrar Grupos

```python
grupos=datos[:,8]
print(grupos)
```

# instanciar PCA (con los componentes) y Transformar los datos 

```python
pca=PCA(n_components=2)
pca.fit(datos[:,:8])
D=pca.transform(datos[:,:8])
```

# Componentes principales 

```python
print(D)
```

# cargar Datos aquellos con prueba negativa

```python
# usar where de numpy para buscar en donde estan los 0 
ceros=np.where(grupos==0)

ejesNX=[x[0] for x in D[ceros]]
ejesNY=[x[1] for x in D[ceros]]
```

# cargar Datos Aquellos con prueba positiva

```python
# usar where de numpy para buscar en donde estan los 1
unos=np.where(grupos==1)
ejesPX=[x[0] for x in D[unos]]
ejesPY=[x[1] for x in D[unos]]

```

# Mostrar la Gr√°fica

```python
fig= plt.figure(figsize=(20,10))
p1=plt.plot(ejesNX,ejesNY,'bo')
p2=plt.plot(ejesPX,ejesPY,'ro')
plt.legend((p1[0], p2[0]), ('Negativos', 'Positivos'))

plt.title("Resultado An√°lisis de Componentes ")
plt.show()
```
